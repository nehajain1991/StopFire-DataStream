# Task A. MongoDB Data Model 

There are two types of models in which the data can be stored in MongoDB – referencing model and embedded model. Below is the pictorial representation of the referencing and embedding data models:<br>
 <br>
As visible from  above, for the embedding model, Document 2 is embedded in Document 1 where the data from Document 2 can be directly pulled in the query while pulling data for Document 2. This model does not support normalization of the data.<br>
In contrast to this, referencing model has no embedding done, we create a link between the two documents through which Document 1 can refer Document 2 data when needed. They are generally used when we want to keep the data normalized stored in different documents.<br><br>

The advantages of reference model are following:<br>
&emsp;  &emsp; 1.	It avoids the repetition of the data stored in the data such as that in the embedding model.<br>
&emsp;  &emsp; 2.	Reduces errors and help in keeping the data consistent.<br>
&emsp;  &emsp; 3.	Reduces the size of the document and remains in the size limit of 16 MB. <br>
&emsp;  &emsp; 4.	Only one document change is required to change data.<br><br>

For the above-mentioned advantages of the referencing model, we have used the Referencing model to store the data both historic and streaming to the MongoDB.  We have one-many relationships between the climate and the hotspot data and it is streaming data, which would go on increasing and never decrease, we want to reduce the repetitions in the data by using embedding model which can increase the size of the document. Additionally, we want to maintain both the climate and the hotspot data separately by creating the link between the for efficient querying of the data. We don't want both the data to interfere with each other.<br><br>

## Task B1. Loading the historic data

Using the above mentioned reference model, we now load the data into file hotspot_historic.csv and climate_historic.csv into the database <b>fit5148_assignment_db</b>.We will create two collections in this database as <b>climate</b> and <b>hotspot</b> and store the data in these two collections respectively.<br>For this, we use the pandas package. We load the files into the pandas dataframe and then load them as json documents in the MongoDB.

```python
#importing the libraries
import pymongo
import pandas as pd
from bson import ObjectId
from pymongo import MongoClient
from pprint import pprint
```

```python
client = MongoClient () # creating Mongo clinet instance
db = client.fit5148_assignment_db # using the database mentioned above
```

```python
climate_historic_df = pd.read_csv("climate_historic.csv") # loading the climate_historic file into dataframe

climate_historic_df.dtypes # checking the data types of the fields
```

```python
#converting the fields into required format to store in MongoDB
climate_historic_df['station'] = climate_historic_df['station'].astype(float) 
climate_historic_df.air_temperature_celcius = climate_historic_df.air_temperature_celcius.astype(float)
```

```python
hotspot_historic_df = pd.read_csv("hotspot_historic.csv") # loading the hotspot_historic file into dataframe

hotspot_historic_df.dtypes # checking the data types of the fields
```

```python
#converting the fields into required format to store in MongoDB
hotspot_historic_df['confidence'] = hotspot_historic_df['confidence'].astype(float)
hotspot_historic_df['surface_temperature_celcius'] = hotspot_historic_df['surface_temperature_celcius'].astype(float)
```

```python
#the two collections that will be used
db.climate 
db.hotspot
```

```python
#loading the data for hotspot and climate into MongoDB using referencing model defined in Task A
for i in climate_historic_df.index:
    c = {
            '_id': str(i),
            'date': climate_historic_df.iloc[i]["date"],
            'station':climate_historic_df.iloc[i]["station"],
            'air_temperature_celcius': climate_historic_df.iloc[i]["air_temperature_celcius"],
            'relative_humidity': climate_historic_df.iloc[i]["relative_humidity"],
            'windspeed_knots': climate_historic_df.iloc[i]["windspeed_knots"],
            'max_wind_speed': climate_historic_df.iloc[i]["max_wind_speed"],
            'precipitation': climate_historic_df.iloc[i]["precipitation "]
    }
    db.climate.insert(c)
    
    
    for j in hotspot_historic_df.loc[hotspot_historic_df["date"] == climate_historic_df.iloc[i]["date"]].index:
        h = {
            '_id': str(j),
            'latitude':hotspot_historic_df.iloc[j]["latitude"],
            'longitude':hotspot_historic_df.iloc[j]["longitude"],
            'datetime':hotspot_historic_df.iloc[j]["datetime"],
            'confidence':hotspot_historic_df.iloc[j]["confidence"],
            'date':hotspot_historic_df.iloc[j]['date'],
            'surface_temperature_celcius':hotspot_historic_df.iloc[j]["surface_temperature_celcius"],
            'climate' : str(i)  # climate is reference here based on the date
        }
        db.hotspot.insert(h)

   



    
```

## Task B2. Querying the loaded data

In this this task, we have a few questions that are needed to be answered using the data loaded in the Task B1. For this we will use the queries to pull data from the collections based on the questions.<br>Please note that the database and collections used are same as in above task

```python
# printing the collections in the database
print(db.list_collection_names())
```

### a. Find climate data on ​ 10th December 2017 ?

```python
x = db.climate.find({'date':{'$eq':'10/12/2017'}})

for document in x: 
    pprint(document)
```

### b. Find   the   latitude ​ ,   ​ longitude,   surface   temperature   ​ (°C)​ ,   ​ and   ​ confidence   ​ when  the surface temperature (°C) was between ​ 65 °C ​ and​ 100 °C ?(assumption:in between we include the 65 and 100)

```python
start = '65'
end = '100'

x = db.hotspot.find({'surface_temperature_celcius':{'$gte':int(start),'$lte':int(end)}},\
                    {'_id':0,'latitude':1,'longitude':1,'surface_temperature_celcius':1,'confidence':1})
for document in x: 
    pprint(document)
```

### c. Find   ​ date , ​   ​ surface   temperature   (°C),   air   temperature   (°C),   relative   humidity   and  max wind speed on ​ 15th and 16th of December 2017 ?

```python
start = '15/12/2017'
end = '16/12/2017'

x = db.climate.aggregate([ {'$lookup' : 
                                     {'from': 'hotspot',
                                      'localField': '_id',
                                      'foreignField': 'climate',
                                      'as': 'results' }},{'$match': {'date':{'$in':[start,end]}}},\
                          {'$project':{'_id':0,'date':1,'air_temperature_celcius':1,
                                                                           'max_wind_speed':1,
                                                                           'relative_humidity':1,
                                                                           'results.surface_temperature_celcius':1}}])
for document in x: 
    pprint(document)
    
```

### d. Find   ​ datetime,   air   temperature   (°C),   surface   temperature   (°C)   and   ​ confidence  when the ​ confidence i ​ s between 80 and 100 ?(assumption: in between we include the 80 and 100 also)


```python
start=80
end=100
x = db.hotspot.aggregate([ {'$lookup' : 
                                     {'from': 'climate',
                                      'localField': 'climate',
                                      'foreignField': '_id',
                                      'as': 'climate' }},{'$match': \
                                    {'confidence':{'$gte':int(start),'$lte':int(end)}}},
                                    {'$project':{'_id':0,'datetime':1,'climate.air_temperature_celcius':1,
                                                                           'confidence':1,
                                                                           'surface_temperature_celcius':1}}])
for document in x:
    pprint(document)
```

### e.Find the top 10 records with the highest ​ surface temperature ​ ​ (°C). ?

```python
x = db.hotspot.aggregate([{'$sort':{'surface_temperature_celcius':-1}}, {'$limit':10}])

for document in x:
        pprint(document)
```

### f. Find   the   number   of   fire   in   each   day.   You   are   required   to   only   display   ​ the  total number of fire ​ and ​ the date ​ in the output

```python
x = db.hotspot.aggregate( [
    {'$group': 
         {'_id':'$date',
         'count': {'$sum' :1 }
        }     
    },
    {
        '$project':{
            '_id': 0, 
           'date': '$_id',
            'Total number of fires': '$count'
        }
    },
    { "$sort": 
           {"Total number of fires" : -1}
             
          }])

for document in x: 
    pprint(document)
    
```

### g.Find   the   ​ average   surface   temperature   ​ (°C)   f ​ or   each   day​ .   You   are   required   to  only display ​ average surface temperature (°C) ​ ​ and ​ the date ​ in the output.?

```python
x = [ 
    {'$group': 
         {'_id':'$date',
         'average': {"$avg" : '$surface_temperature_celcius' }
        }},
    {
        '$project':{
            '_id': 0, 
           'date': '$_id',
            'Average surface temperature': '$average'
        }},
        { "$sort": 
           {"Average surface temperature" : -1}
             
          }
    ]

for document in db.hotspot.aggregate(x): 
    pprint(document)
```
